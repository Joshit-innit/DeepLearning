```python
import torch
import time
```

```python


def matrix_addition(a, b):
    # CPU implementation
    start_time = time.time()
    c = a + b
    end_time = time.time()
    cpu_time = end_time - start_time
    print(f"CPU time: {cpu_time: .6f} seconds")

    # GPU implementation
    a_gpu = a.to("mps")
    b_gpu = b.to("mps")
    start_time = time.time()
    c_gpu = a_gpu + b_gpu
    torch.mps.synchronize() # wait for the MPS to finish
    end_time = time.time()
    gpu_time = end_time - start_time
    print(f"GPU time: {gpu_time: .6f} seconds")

```
